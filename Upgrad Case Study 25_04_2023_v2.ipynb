##Clustering - Model Building 

### Data Preparation
It is done based on - 
Avg payment time in days and standard dev for payment time

Two derived variables: the average payment time in days for a customer and the standard deviation for the payment time.
Using clustering techniques would result in a few distinct clusters of customers, which can be used as an input variable for the ML model.

```python
# Subtracting to find delta days
rec_df['pay_delta_days'] = (rec_df['RECEIPT_DATE'] - rec_df['INVOICE_CREATION_DATE']).astype(str)
# Removing days and changing to integer
rec_df['pay_delta_days'] = rec_df['pay_delta_days'].apply(lambda x : x.replace(' days',''))

rec_df['pay_delta_days'] = rec_df['pay_delta_days'].astype(int)

rec_df['pay_delta_days'].head()
```

```python
 # Finding term by subtracting due date from invoice creation date
    
rec_df['Term_days'] = (rec_df['DUE_DATE'] - rec_df['INVOICE_CREATION_DATE']).astype(str)

rec_df['Term_days'] = rec_df['Term_days'].apply(lambda x : x.replace(' days',''))

rec_df['Term_days'] = rec_df['Term_days'].astype(int)

rec_df['Term_days'].head()
```

```python
# Grouping by customer name and creating dataframe that will include average and std for pay_delta

rec_df_cluster_days = rec_df.groupby('CUSTOMER_NAME')['pay_delta_days'].aggregate(['mean','std'])

rec_df_cluster_days.columns= ['days_mean','days_std']

rec_df_cluster_USD = rec_df.groupby('CUSTOMER_NAME')['USD Amount'].aggregate(['mean','std'])

rec_df_cluster_USD.columns= ['USD_mean','USD_std']

rec_df_cluster_term = rec_df.groupby('CUSTOMER_NAME')['Term_days'].aggregate(['mean','std'])
rec_df_cluster_term.columns= ['term_mean','term_std']

rec_df_clusters = rec_df_cluster_days.join(rec_df_cluster_USD).join(rec_df_cluster_term)

rec_df_clusters
```

### Hopkins Test

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

A value close to 1 tends to indicate the data is highly clustered, random data will tend to result in values around 0.5, and uniformly distributed data will tend to result in values close to 0.
